```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
file = '/content/drive/MyDrive/Colab Notebooks/coding_dojo/files/Data.xlsx'
df = pd.read_excel(file)
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
uniqueSports = df.groupby(['Sport'])[2004].unique()
sport = uniqueSports.index
count = uniqueSports.values
plt.xticks(rotation=90)
plt.plot(sport, count, marker='.', markersize=10, c='red')
plt.grid(axis='x')
plt.title("Popular Sports in 2004", fontsize=20)
plt.xlabel("Sports", fontsize=16)
plt.ylabel("Count", fontsize=16)
# Most popular sport is definietly football
```

```python
uniqueSports = df.groupby(['Sport'])[2017].unique()
sport = uniqueSports.index
count = uniqueSports.values
plt.xticks(rotation=90)
plt.plot(sport, count, marker='.', markersize=10, c='purple')
plt.grid(axis='x')
plt.title("Popular Sports in 2004", fontsize=20)
plt.xlabel("Sports", fontsize=16)
plt.ylabel("Count", fontsize=16)
# Most popular sport is still football
# None, figure skating, soccer, volleyball have gone up in sports
# baseball and basketball have both seen drops
```

```python
plt.figure(figsize=(18,10))
uniqueSports = df.groupby(['Sport'])[2004].unique()
uniqueSports05 = df.groupby(['Sport'])[2005].unique()
uniqueSports06 = df.groupby(['Sport'])[2006].unique()
uniqueSports07 = df.groupby(['Sport'])[2007].unique()
uniqueSports08 = df.groupby(['Sport'])[2008].unique()
uniqueSports13 = df.groupby(['Sport'])[2013].unique()
uniqueSports17 = df.groupby(['Sport'])[2017].unique()

plt.xticks(rotation=90)
plt.plot(uniqueSports.index, uniqueSports.values, marker='.', markersize=10, c='red')
plt.plot(uniqueSports05.index, uniqueSports05.values, marker='.', markersize=10, c='blue')
plt.plot(uniqueSports06.index, uniqueSports06.values, marker='.', markersize=10, c='purple')
plt.plot(uniqueSports07.index, uniqueSports07.values, marker='.', markersize=10, c='yellow')
plt.plot(uniqueSports08.index, uniqueSports07.values, marker='.', markersize=10, c='orange')
plt.plot(uniqueSports13.index, uniqueSports07.values, marker='.', markersize=10, c='gray')
plt.plot(uniqueSports17.index, uniqueSports07.values, marker='.', markersize=10, c='black')
plt.grid(axis='x')
plt.title("Popular Sports in 2004", fontsize=20)
plt.xlabel("Sports", fontsize=16)
plt.ylabel("Count", fontsize=16)
plt.legend(loc="center right")
# Most popular sport is definietly football
# looks like year over year there's some flucuation
```
