```python
import pandas as pd
```

```python
# Uploading from a URL
filename = 'https://raw.githubusercontent.com/mGalarnyk/Tutorial_Data/master/Pandas/CalIt2.data'
df = pd.read_csv(filename, header = None)
```

## Renaming
1. Rename column 0 to 'Flow_ID'
2. Rename column 1 to 'Date'
3. Rename column 2 to 'Time'
4. Rename column 3 to 'Count'

```python
df = df.rename(columns={0: 'Flow_ID', 1: 'Date', 2: 'Time', 3: 'Count'})
```

```python
df.head()
```

## Selecting Data
1. Select all data for the date July 24 2005 having flow id=7.
2. From the data selected in step 1, sort the result on Count in descending order and output the top 10 rows. Assign the output to a variable named top10.

```python
# 1. Select all data for the date July 24 2005 having flow id=7.
july = df['Date'] == '07/24/05'
flow = df['Flow_ID'] == 7;

# == different tries == 
# df[july & flow] - No difference between this and on th eone below
# july_flow = df.loc[july & flow,:]
# july_flow = july & flow

july_flow = (df['Date'] == '07/24/05') & (df['Flow_ID'] == 7)
july_flow.head()
```

```python
# 2. From the data selected in step 1, sort the result on Count in descending order and output the top 10 rows. Assign the output to a variable named top 10.
top_ten = df.loc[july_flow,:].sort_values(by='Count', ascending=False).head(10)
top_ten
```

## Apply Function
1. For the 10 rows outputted above, use Pandas Apply function to subtract lowest value of the 10 from all of them and then output the average value of the resulting counts

```python
# lawest value
min_val = top_ten['Count'].min()
```

```python
# sum minus the lowest value
def lowest_average(x):
  return x - min_val
```

```python
# Finding the mean
new_data = top_ten['Count'].apply(lowest_average)
# new_data= lowest_average(range['Count'])  
new_data.mean()
```

## Grouping
1. Select data in the month of August 2005 having flow id=7
2. Group the data based on date and get the max count per date

```python
# 1. Select data in the month of August 2005 having flow id=7
august_start = df['Date'].str.startswith('07')
august_end = df['Date'].str.endswith('05')
flow = df['Flow_ID'] == 7

august_flow = df.loc[august_start & august_end & flow,:]

august_flow.head()
```

```python
# 2. Group the data based on date and get the max count per date
max_date = df.groupby(['Date']).max()
max_date.head()
```
